**# Hypothesis Testing**

```python
#Menggabungkan Fitur Kolom Education dan Income untuk menganalisis Uji Hipotesis (cara 1)
Edu_Income = df.groupby(['Education'])['Income']
Edu_Income.describe()
```

```python
df.groupby(['Education'])['Income'].describe() # (cara 2)
```

```python
#Metode Pengujian Hypothesis menggunakan Anove

dfCycle = df[df['Education'] == '2n Cycle']
dfBasic = df[df['Education'] == 'Basic']
dfGraduation = df[df['Education'] == 'Graduation']
dfMaster = df[df['Education'] == 'Master']
dfPhD = df[df['Education'] == 'PhD']
```

```python
anove_test = stats.f_oneway(dfCycle['Income'],dfBasic['Income'],dfGraduation['Income'],dfMaster['Income'],dfPhD['Income'])
```

```python
result = anove_test.pvalue
```

```python
if result > 0.5 :
  print('Education tidak berpengaruh terhadap Income')
else:
  print('Education berpengaruh terhadap Income')
```

Hasil Hypothesis Testing menunjukkan bahwa Education berpengaruh terhadap Income
